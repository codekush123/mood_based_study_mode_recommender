```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
data = pd.read_csv('mood_data.csv', sep=';')
data.head()
```

```python
features = [
    "user_energy_state",
    "user_emotional_state",
    "available_time_minutes",
    "day_type",
    "last_session_outcome",
    "recent_study_duration_avg"
]
target = "recommended_study_mode"
```

```python
data = data.dropna(subset=[target])
```

```python
encoders = {}
X = pd.DataFrame()
```

```python
for col in features:
    if data[col].dtype == 'object':
        le = LabelEncoder()
        X[col] = le.fit_transform(data[col])
        encoders[col] = le
    else:
        X[col] = data[col]
```

```python
target_encoder = LabelEncoder()
y = target_encoder.fit_transform(data[target])
```

```python
train_model = DecisionTreeClassifier(max_depth=5,random_state=42)
train_model.fit(X, y)
```

```python
joblib.dump(train_model, "mood_based_study_mode_model.pkl")
joblib.dump(encoders, "feature_encoders.pkl")
joblib.dump(target_encoder, "target_encoder.pkl")
```
